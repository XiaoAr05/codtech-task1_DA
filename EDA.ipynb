```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
sales_data = pd.read_csv('sales_data.csv')

# Display the first 5 rows of the dataset
print("First 5 rowsof the dataset:\n", sales_data.head())
```

```python
print("Shape of the dataset:", sales_data.shape) 
```

```python
print("Data types:\n", sales_data.dtypes)
```

```python
print("Summary statistics:\n", sales_data.describe())
```

```python
print("Missing values:\n", sales_data.isnull().sum())
```

```python
# Total Sales Distribution
sales_data['Total_Sales'].hist(bins=10, color='skyblue')
plt.title('Total Sales Distribution')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Units Sold
sales_data['Units_Sold'].hist(bins=10, color='lightgreen')
plt.title('Units Sold Distribution')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Product', data=sales_data, color='brown')
plt.title('Product Count Distribution')
plt.xlabel('Product')
plt.ylabel('Count')
plt.show()
```

```python
# Total Sales by Product
product_sales = sales_data.groupby('Product')['Total_Sales'].sum()
product_sales.plot(kind='bar', color='orange')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Units Sold Over Time
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data.sort_values('Date', inplace=True)
plt.plot(sales_data['Date'], sales_data['Units_Sold'], marker='o', linestyle='-', color='green')
plt.title('Units Sold Over Time')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
correlation = sales_data[['Units_Sold', 'Unit_Price', 'Total_Sales']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter Plot for Units Sold vs. Total Sales
plt.scatter(sales_data['Units_Sold'], sales_data['Total_Sales'], color='purple')
plt.title('Scatter Plot: Units Sold vs. Total Sales')
plt.xlabel('Units Sold')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Scatter Plot for Unit Price vs. Total Sales
plt.scatter(sales_data['Unit_Price'], sales_data['Total_Sales'], color='blue')
plt.title('Scatter Plot: Unit Price vs. Total Sales')
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
sns.boxplot(x=sales_data['Total_Sales'], color='red')
plt.title('Outliers in Total Sales')
plt.xlabel('Total Sales')
plt.show()
```

```python
# Pair plot for numerical features
sns.pairplot(sales_data, vars=['Units_Sold', 'Unit_Price', 'Total_Sales'], hue='Product')
plt.show()
```

```python
sns.swarmplot(x=sales_data['Total_Sales'], color='purple')
plt.title('Swarm Plot for Outliers in Total Sales')
plt.xlabel('Total Sales')
plt.show()
```
